```python
# Importando bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Definindo meu dataframe
# O dataframe carregado traz informações sobre a participação das mulheres no mercado de trabalho ao longo dos anos 70 e 2012
url = 'https://raw.githubusercontent.com/brunadejesusramos/Michigan_charts/master/USALFPWNA.csv'
df = pd.read_csv(url)
df['participation_evolution']=df['USALFPWNA'].diff()
df.fillna(0,inplace=True)
```

```python
ax1=df.plot(x='DATE', y='USALFPWNA', kind="line",color='g',figsize=(10,5))
ax1.set_ylabel('USA Women Rate',color='g')


ax2 = df['participation_evolution'].plot(secondary_y=True, color='k', marker='o')
ax2.set_ylabel('% participation_evolution',color='k')


ax1.set_xticklabels(df.DATE, rotation=90,)
ax1.set_xticks(np.arange(len(df.DATE)))
plt.legend(['% participation_evolution'], loc=9)
```

Com a analise do grafico é possivel identificar que dos anos 70 até o inicio dos anos 90 houve um grande crescimento na participação das mulheres no mercado de trabalho. A evolução do crescimento teve oscilações de queda a partir de 1975, esse ano foi seguido de um pico de crescimento em 1978, depois desse ano houve uma queda acentuada no volume de evolução de participação.
Embora o volume de evolução participação tenha oscilado entre os anos o percentual de participação manteve-se sempre crescente.
